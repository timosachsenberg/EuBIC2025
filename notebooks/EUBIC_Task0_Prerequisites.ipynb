## 1.5 Functions

Reusable blocks of code.

````python
def calculate_mz(mass, charge):
    """
    Calculate m/z from mass and charge.
    
    Parameters:
    -----------
    mass : float
        Neutral mass in Daltons
    charge : int
        Charge state (positive)
    
    Returns:
    --------
    float : m/z value
    """
    proton_mass = 1.00728
    return (mass + charge * proton_mass) / charge

# Use the function
print(f"m/z of 1000 Da at z=1: {calculate_mz(1000, 1):.4f}")
print(f"m/z of 1000 Da at z=2: {calculate_mz(1000, 2):.4f}")
print(f"m/z of 1000 Da at z=3: {calculate_mz(1000, 3):.4f}")
````

````python
# Functions with default parameters
def calculate_mz(mass, charge=2):  # charge defaults to 2
    proton_mass = 1.00728
    return (mass + charge * proton_mass) / charge

print(f"Default z=2: {calculate_mz(1000):.4f}")
print(f"Override z=3: {calculate_mz(1000, charge=3):.4f}")
````

## 4.2 Mass-to-Charge Ratio (m/z)

The **m/z** is the fundamental measurement in mass spectrometry.

```
         mass (in Daltons)
m/z = â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
         charge (integer)
```

**Example:** A peptide with mass 1000 Da:
- At charge +1: m/z = 1000/1 = 1000
- At charge +2: m/z = 1000/2 = 500
- At charge +3: m/z = 1000/3 = 333.3

**Important:** The mass spectrometer measures m/z, not mass directly. You must know the charge to calculate the mass!

````python
# m/z calculation (including proton mass for ESI)
def mass_to_mz(mass, charge):
    """Convert neutral mass to m/z for ESI."""
    proton_mass = 1.00728
    return (mass + charge * proton_mass) / charge

def mz_to_mass(mz, charge):
    """Convert m/z back to neutral mass."""
    proton_mass = 1.00728
    return (mz * charge) - (charge * proton_mass)

# Example
neutral_mass = 1000
for z in [1, 2, 3]:
    mz = mass_to_mz(neutral_mass, z)
    print(f"Mass {neutral_mass} Da at z={z}: m/z = {mz:.2f}")
````

# 5. File Formats

Common file formats you'll encounter in proteomics.

## 5.1 FASTA Format

Text format for protein/DNA sequences:

```
>sp|P02769|ALBU_BOVIN Serum albumin OS=Bos taurus
MKWVTFISLLLLFSSAYSRGVFRRDTHKSEIAHRFKDLGEEHFKGLVLIAFSQYLQQCPF
DEHVKLVNELTEFAKTCVADESHAGCEKSLHTLFGDELCKVASLRETYGDMADCCEKQEP
ERNECFLSHKDDSPDLPKLKPDPNTLCDEFKADEKKFWGKYLYEIARRHPYFYAPELLYY
...
```

**Structure:**
- Header line starts with `>`
- Contains identifier, name, and metadata
- Sequence follows on subsequent lines
- One-letter amino acid codes

## 5.2 mzML Format

Open XML format for raw mass spectrometry data:

```xml
<mzML>
  <spectrumList count="10000">
    <spectrum index="0" id="scan=1">
      <cvParam name="ms level" value="1"/>  <!-- MS1 -->
      <cvParam name="scan start time" value="10.5"/>  <!-- RT -->
      <binaryDataArrayList>
        <binaryDataArray>  <!-- m/z values -->
          ...
        </binaryDataArray>
        <binaryDataArray>  <!-- intensity values -->
          ...
        </binaryDataArray>
      </binaryDataArrayList>
    </spectrum>
    ...
  </spectrumList>
</mzML>
```

**Contains:**
- All spectra (MS1 and MS2)
- m/z and intensity arrays
- Retention time
- Precursor information for MS2

## 5.3 OpenMS File Formats

| Format | Extension | Contains |

# 6. pyOpenMS Object Model

pyOpenMS provides Python bindings to OpenMS, a C++ library for computational mass spectrometry.

````python
import pyopenms as oms
print(f"pyOpenMS version: {oms.__version__}")
````

## 6.1 Core Data Types

````python
# AASequence - represents a peptide
peptide = oms.AASequence.fromString("PEPTIDER")

print(f"Sequence: {peptide}")
print(f"Length: {peptide.size()} amino acids")
print(f"Monoisotopic mass: {peptide.getMonoWeight():.4f} Da")
print(f"Average mass: {peptide.getAverageWeight():.4f} Da")
````

````python
# MSSpectrum - represents a single spectrum
spectrum = oms.MSSpectrum()

# Add peaks (m/z, intensity)
spectrum.push_back(oms.Peak1D(500.0, 1000.0))
spectrum.push_back(oms.Peak1D(600.0, 2000.0))
spectrum.push_back(oms.Peak1D(700.0, 1500.0))

# Set metadata
spectrum.setRT(120.5)  # retention time in seconds
spectrum.setMSLevel(1)  # MS1

print(f"Number of peaks: {spectrum.size()}")
print(f"RT: {spectrum.getRT()} s")
print(f"MS level: {spectrum.getMSLevel()}")
````

````python
# Access peaks as numpy arrays
mzs, intensities = spectrum.get_peaks()
print(f"m/z values: {mzs}")
print(f"Intensities: {intensities}")
````

````python
# MSExperiment - container for multiple spectra
exp = oms.MSExperiment()

# Add spectra
for i in range(3):
    spec = oms.MSSpectrum()
    spec.setRT(100 + i * 10)
    spec.setMSLevel(1)
    spec.push_back(oms.Peak1D(500 + i * 10, 1000.0))
    exp.addSpectrum(spec)

print(f"Number of spectra: {exp.size()}")
print(f"RT range: {exp[0].getRT()} - {exp[-1].getRT()} s")
````

## 6.3 Working with Algorithms

````python
# Most algorithms follow this pattern:

# 1. Create algorithm instance
tsg = oms.TheoreticalSpectrumGenerator()

# 2. Get and modify parameters
params = tsg.getParameters()
params.setValue("add_b_ions", "true")
params.setValue("add_y_ions", "true")
tsg.setParameters(params)

# 3. Run the algorithm
peptide = oms.AASequence.fromString("PEPTIDER")
theo_spectrum = oms.MSSpectrum()
tsg.getSpectrum(theo_spectrum, peptide, 1, 2)  # charge 1 to 2

print(f"Generated {theo_spectrum.size()} theoretical peaks")
````
